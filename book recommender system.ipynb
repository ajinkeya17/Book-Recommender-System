We need a pivot table where one axis has all books and other has Users and the values are the ratings.
To build an intelligent recommendor system we will consider only those Users who have reviewed atleast 200 books so that we only include ratings made by well-read people and we include only those books which have been reviewed by atleast 50 people so that there are only popular books in our roster.

```python
# Filtering out users who have reviewed atleast 200 books
temp = ratings_with_name.groupby('User-ID')['Book-Rating'].agg('count')
temp
```

```python
temp[temp>200].index
```

```python
df = ratings_with_name[ratings_with_name['User-ID'].isin(temp[temp>200].index)]
df
```

```python
# Applying pivot table on this df
pt = df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

Now each book is a 810 dim vector. We compute the distance metric(euclidian,manhattan,angular dist = cosine sim,etc) between each book vector with all other book vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(pt)
```

```python
similarity  # first array is similarith score of the book 1984 with all other books. first value is 1 as it is the similarity of the book with itself and hence all diagonal elements are 1
```

# Creating the recommend function
It will take a book as input and return 5 suggestions

```python
def recommend(book):
    # retrieve index of book name from pivot table
    ind = np.where(pt.index==book)[0][0]
    sim_items = sorted(list(enumerate(similarity[ind])), key = lambda x : x[1], reverse=True)[1:6]
    for item in sim_items:
        print(pt.index[item[0]])
```

```python
recommend('The Hobbit : The Enchanting Prelude to The Lord of the Rings')
```
